<p style="font-size:30px; color:green; text-align:centre "><b>K-means Clustering</b></p> 

```python
# Plot the graph to show the relations
import matplotlib.pyplot as plt


df.plot(kind = 'scatter', x = 'annual_base_pay', y = 'job_cat_n', xlim = (0, 300000))
```

```python
# Create the target columns dataframe

data = df[['annual_base_pay','job_cat_n',]]

print(data)
```

```python
#Elbow Method


from sklearn.preprocessing import StandardScaler

scaled_df = StandardScaler().fit_transform(data)

from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
kmeans_kwargs = {
    "init" : "random",
    "n_init": 10,
    "random_state": 1
}

#Create list to hold SSE values for each k

sse = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)
    
#visualize results
plt.plot(range(1,11), sse)
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')

plt.show()
```

```python
#Create a Kmeans clustering object with 3 clusters
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

kmeans = KMeans(n_clusters=3)


#Fit the data to the KMeans model
kmeans.fit(data)

#Get the cluster labels for each data point
labels = kmeans.labels_
cetroids = kmeans.cluster_centers_

#Plot the results 
plt.scatter(data['annual_base_pay'], data['job_cat_n'], c=labels, cmap='brg')
plt.scatter(cetroids[:,0], cetroids[:,1], marker='x', c='k', s=200)
plt.xlabel('Annual Base Pay')
plt.ylabel('Job Category')
plt.title('k-means clustering of Salary Survey Result for k=3')
plt.show()
```

```python
#Create new column on the basis of salary criteria mentioned below

df.loc[df['annual_base_pay'] >= 100000, 'salary_more_than_100k'] = 1

df.loc[df['annual_base_pay'] < 100000, 'salary_more_than_100k'] = 0
```

```python
#Create the input dataframe and drop the target column from the original dataframe 

inputs = df.drop('salary_more_than_100k',axis='columns')
```

```python
#Create the target dataframe 

target = df['salary_more_than_100k']
```

```python
df['salary_more_than_100k'].value_counts()
```

```python
inputs_n = df[['employer_n', 'country_n', 'job_cat_n']]
```

```python
target
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python


#Split your dataset into training and testing sets:
X_train1, X_test1, y_train1, y_test1 = train_test_split(inputs_n, target, test_size=0.3, random_state=42)
```

```python
import matplotlib.pyplot as plt

#Tree visualization
fig, ax = plt.subplots(figsize=(10,10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

```python
#Train decision tree model:


clf = tree.DecisionTreeClassifier()

clf.fit(X_train1, y_train1)
```

```python
#Evaluate the model:
y_pred = clf.predict(X_test1)
print(classification_report(y_test1, y_pred))
```

<b style="font-size:20px">Resolving imbalance in data with SMOTE</b>

```python
from imblearn.over_sampling import SMOTE



smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train1, y_train1)
```

```python
print(y_train_resampled.value_counts())
```

```python
import matplotlib.pyplot as plt

class_count_updated = y_train_resampled.value_counts()

class_count_updated

plt.figure(figsize=(6,4))
plt.pie(class_count_updated, labels=class_count_updated.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Classes after Balancing Data')
plt.show()
```

```python
#Implementing the SVM model

from sklearn.svm import SVC
from sklearn.metrics import classification_report

svm_model = SVC(kernel='linear', random_state=100)
svm_model.fit(X_train_resampled, y_train_resampled)

y_pred = svm_model.predict(X_test1)


#Evaluate the model
print(classification_report(y_test1, y_pred))
```

```python
#Implementing the Decision Tree model for the resampled data

#Train decision tree model:


clf_updated = DecisionTreeClassifier()

clf_updated.fit(X_train_resampled, y_train_resampled)

#Evaluate the model:
y_pred_updated= clf.predict(X_test1)
print(classification_report(y_test1, y_pred_updated))
```

```python
clf_updated.score(X_test1,y_test1)
```
